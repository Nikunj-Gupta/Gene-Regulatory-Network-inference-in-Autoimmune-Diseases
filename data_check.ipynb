```python
import GEOparse 
import pandas as pd, numpy as np 
import glob, sys 
from pprint import pprint 
```

```python
datasets_path = './datasets/soft_files' 
datasets = glob.glob(datasets_path+'/*') 
pprint(datasets)
```

```python
shortlisted_datasets = [0, 1, 3, 5, 6] # Datasets with Gene Symbol column and ELF1 and SYNGR1 
# shortlisted_datasets = [0, 3, 5, 6] # Datasets with Gene Symbol column and ELF1 and SYNGR1 
datasets = [datasets[i] for i in shortlisted_datasets] 
pprint(datasets)
```

```python
# Checking for ELF1 and SYNGR1 in platform (GPL)
id_refs = {} 
for data in datasets: 
    print("================================================================================================") 
    print(data.split('/')[-1]) 
    id_refs[data.split('/')[-1]] = {"ELF1": None, "SYNGR1":None} 
    gse = GEOparse.get_GEO(filepath=data, silent=True) 
    for gpl_name, gpl in gse.gpls.items():
        print("Name: ", gpl_name)
        # print("Metadata:",)
        # for key, value in gpl.metadata.items():
        #     print(" - %s : %s" % (key, ", ".join(value)))
        # print("Table data:",)
        # print(gpl.table) 
        # print(gpl.table.columns)
        break 
    print("======== ELF1 ========") 
    print(gpl.table[gpl.table['Gene Symbol']=='ELF1']) 
    id_refs[data.split('/')[-1]]["ELF1"] = list(gpl.table[gpl.table['Gene Symbol']=='ELF1']['ID']) 
    print("======== SYNGR1 ========") 
    print(gpl.table[gpl.table['Gene Symbol']=='SYNGR1']) 
    id_refs[data.split('/')[-1]]['SYNGR1'] = list(gpl.table[gpl.table['Gene Symbol']=='SYNGR1']['ID']) 
    print("================================================================================================") 
    print() 
    # break 
pprint(id_refs)
```

```python
def check(List1, List2): 
    if any(item in List1 for item in List2): 
#         print("The list1 contains some elements of the list2") 
        return True 
    else: 
#         print("No, List1 doesn't have any elements of the List2.") 
        return False 
```
