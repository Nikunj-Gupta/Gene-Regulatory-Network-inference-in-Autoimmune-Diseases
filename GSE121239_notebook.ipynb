```python
def load(data_dir='datasets/GSE121239/GSE121239/'): 
    raw_expre = pd.read_csv(data_dir+'/raw_expre.csv')  
    log2_expre = pd.read_csv(data_dir+'/log2_expre.csv')  
    pheno = pd.read_csv(data_dir+'/pheno.csv') 
    return raw_expre, log2_expre, pheno 
```
