```python
def sort_df(df, sort_col='adj.P.Val'): 
    #Sort ascending or descending (depending on sort_col) 
    if sort_col in ['adj.P.Val']: 
        if df[sort_col].is_monotonic_increasing: return df 
        else: return df.sort_values(by=sort_col, ascending=True).reset_index(drop=True) 
    elif sort_col in ['t', 'B', 'logFC']: 
        if df[sort_col].is_monotonic_decreasing: return df 
        else: return df.sort_values(by=sort_col, ascending=False).reset_index(drop=True) 
```

```python
res_df_cols = ['Dataset', 'sort_by', 'rank_elf1', 'rank_irf1', 'rank_syngr1'] 
res_df = pd.DataFrame(columns=res_df_cols) 

for data in datasets: 
    df = pd.read_csv(data) 
    name = data.split('/')[-1].upper()
    for col in ['adj.P.Val', 't', 'B', 'logFC']:
        sorted_df = sort_df(df, sort_col=col) 
        res_df.loc[len(res_df.index)] = [name, col, 
                                         list(sorted_df[sorted_df['Gene.symbol']=='ELF1'].index)[0], 
                                         list(sorted_df[sorted_df['Gene.symbol']=='IRF1'].index)[0], 
                                         list(sorted_df[sorted_df['Gene.symbol']=='SYNGR1'].index)[0]]
res_df 
```
